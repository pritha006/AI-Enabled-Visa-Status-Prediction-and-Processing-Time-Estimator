```python
import pandas as pd

df = pd.read_csv('Visadataset.csv')
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
df['case_status'].value_counts()
```

```python
# Handle missing values
df = df.dropna()

# Encode categorical features
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
X1 = df.drop('case_status', axis=1)
y1 = df['case_status']

from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=200, random_state=42)
clf.fit(X1_train, y1_train)

y1_pred = clf.predict(X1_test)
import joblib

joblib.dump(clf, "visa_status_model.pkl")
print("Visa status model saved as visa_status_model.pkl")

```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y1_test, y1_pred))
print(classification_report(y1_test, y1_pred))
```

```python
X2 = df.drop('unit_of_wage', axis=1)
y2 = df['unit_of_wage']

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(n_estimators=300, random_state=42)
reg.fit(X2_train, y2_train)

y2_pred = reg.predict(X2_test)
import joblib

# Save model
joblib.dump(reg, "processing_time_model.pkl")
print("Model saved as processing_time_model.pkl")


```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
print("MAE: ", mean_absolute_error(y2_test, y2_pred))
print("RMSE: ", np.sqrt(mean_squared_error(y2_test, y2_pred)))
```
